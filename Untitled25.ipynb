DS - Ensemble method (bagging)

```python
#bagging with decision Tree
#import libraries
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#synthics data
x, y = make_classification(
    n_samples=1000,
    n_features=10,
    random_state = 20
)
print(x,y)


```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=20)
```

```python
#Intialize the decision Tree Classifier
base_estimator = DecisionTreeClassifier(random_state=20)
```

```python
#Initialize the baggin classifier
bagging_model = BaggingClassifier(estimator = base_estimator,n_estimators = 15,random_state = 20)
```

```python
#Train the bagging classifier
bagging_model.fit(x_train,y_train)
```

```python
#make prediction
y_pred = bagging_model.predict(x_test)
```

```python
#evalute model
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```
